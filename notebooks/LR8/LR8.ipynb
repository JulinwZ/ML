Подключение необходимых библиотек

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import BaggingClassifier
from sklearn.metrics import classification_report
```

##Разбиение признаков на X и y, а также на обучающую и тестовую выборку. X - данные, а y - целевой признак. Размер тестовой выборки составляет 15 процентов.

```python
x = dataset.drop(["machine_status"], axis = 1)
y = dataset["machine_status"]
```

##Балансировка и новое разбиение на обучающую и тестовую выборку

Поскольку мы помним о дисбалансе классов, попробуем сбалансировать классы с помощью UnderSampe

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.15)
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
from imblearn.under_sampling import RandomUnderSampler
undersample = RandomUnderSampler(random_state=42)
x_under, y_under = undersample.fit_resample(X_train, y_train)
x_under.shape, y_under.shape, X_test.shape, y_test.shape
```

Как видим, результат не является лучшим, поскольку данных крайне мало и модели не обучатся в таком случае, тогда попробуем OverSample




```python
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state=42)
X_resampled, y_resampled = ros.fit_resample(x, y)
X_resampled.shape, y_resampled.shape
```

Создадим заново обучающую и тестовую выборку с уже нормальной балансировкой классов

```python
X_resampled, X_test, y_resampled, y_test = train_test_split(X_resampled, y_resampled, test_size=0.15)
X_resampled.shape, y_resampled.shape, X_test.shape, y_test.shape
```

Такой вариант подходит лучше

##Понижение размерности и отбор признаков

Создадим функцию, благодаря которой будем проверять, сколько признаков является оптимальным вариантом и на сколько точные предсказания. Помимо этого посмотрим, сколько признаков можно оставить, чтобы от этого не упало качество моделей. Проверять будем на ансамблевых моделях.

```python
def testC(x, Y):
  X_train, X_test, y_train, y_test = train_test_split(x, Y, test_size=0.15, shuffle=True)
  bag = BaggingClassifier().fit(X_train, y_train)
  print(classification_report(y_test, bag.predict(X_test)))
```

Использование метода главных компонент для попытки понижения размерности

```python
from sklearn.decomposition import PCA
for i in range(1,6):
    pca = PCA(n_components=i)
    X_pca = pca.fit_transform(X_resampled, y_resampled)
    print(X_pca.shape)
    print(sum(pca.explained_variance_ratio_))
    testC(X_pca, y_resampled)
    print("-------------")
```

Одномерный отбор признаков

```python
from sklearn.feature_selection import SelectKBest
for i in range(1,8):
    print("count_of_features=",i)
    skb = SelectKBest(k=i)
    X_skb = skb.fit_transform(X_resampled, y_resampled)
    testC(X_skb, y_resampled)
    print(skb.get_feature_names_out())
```

Как мы видим, оптимальным вариантом по выбору признаков является набор всего из трех признаков, хотя их в изначальном датасете 53. Понижение размерности целесообразно делать, поскольку в таком случае времени на вычисления будет затрачено меньше и потенциально модель сможет быстрее находить результат. Тогда сделаем датасет всего из трех представленных признаков

##Создание моделей классификации

Подключение библиотек для моделей классификации:

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
```

Модель логистической регрессии:

```python
from sklearn.linear_model import LogisticRegression


clf = LogisticRegression(random_state=42).fit(X_itog, y_resampled)
y_lr_pred = clf.predict(X_itog_test)
print(classification_report(y_test, y_lr_pred))
```

На данном отчете показано в поле accuracy - точность прдсказаний. Macro avg отвечает за точность предсказаний в целом по разным классам. Ниже представленная confusion matrix - это матрица путаниц. По главной диагонали расположены правильно предсказанные объекты. А на других местах то, сколько объектов было неверно отнесено к другому классу.

```python
print(accuracy_score(y_test, y_lr_pred))
print(confusion_matrix(y_test, y_lr_pred))
```

Модель наивного Байеса

```python
gnb = GaussianNB()
gnb.fit(X_itog, y_resampled)
y_nb_pred = gnb.predict(X_itog_test)
print(classification_report(y_test, y_nb_pred))
```

```python
print(accuracy_score(y_test, y_nb_pred))
print(confusion_matrix(y_test, y_nb_pred))
```

```python
%%writefile app.py
import joblib
import pandas as pd
import numpy as np
import streamlit as st
import matplotlib.pyplot as plt
import seaborn as sns


def main():
    model_1 = load_model("/content/drive/MyDrive/model_LogReg.h5")
    test_data = load_test_data("/content/drive/MyDrive/preproceed_pump.csv")
    data = load_data("/content/drive/MyDrive/preproceed_pump.csv")

    page = st.sidebar.selectbox(
        "Выберите страницу",
        ["Информация о датасете", "Информация о построенных моделях", "Илюстрация зависимостей", "Запрос к модели"]
    )

    if page == "Информация о датасете":
        st.title("Описание задачи и данных")
        st.write("Для перехода на иную страницу можете выбрать ее слева")

        st.header("Описание задачи")
        st.markdown("""Набор данных содержит информацию, записанную с различных сенсоров. Также в датасете имеется два столбца month-месяц, когда были собраны данные, и machine_status  - содержащий информацию о работоспособности насоса""")
        st.write(data.iloc[:5, :])

        st.header("Описание данных")
        st.markdown("""Предоставленные числовые признаки:
* sensor_00-sensor_51 - показания с сенсоров, которые записаны в числовом варианте
* month - число месяца собранных данных,
Целевой признак:
* machine_status - в каком состоянии находится на данный момент насос. Где 0 - нормальное состояние насоса, 1 - восстанавливающийся, 2 - сломанный.

При изучении данного датасета оказалось, что можно оставить три наиболее важных признака, исходя из значений которых, почти с полной уверенностью можно определить значение целевого признака. Эти три признака: sensor04, sensor10, sensor12.
""")
    elif page == "Илюстрация зависимостей":
        st.title("Илюстрация зависимостей")
        st.write("Для перехода на иную страницу можете выбрать ее слева")
        request = st.selectbox("Выберите запрос",["Диаграмма корреляции", "Диаграмма рассеивания целевого признака от остальных", "Диаграмма рассеивания двух признаков", "Круговая диаграмма"])
        if request == "Диаграмма корреляции":
            fig, ax = plt.subplots(figsize=(20,20), dpi= 60)
            sns.heatmap(data.corr(), ax=ax, annot = True)
            st.write(fig)
        elif request == "Диаграмма рассеивания целевого признака от остальных":
            fig, axes = plt.subplots(nrows=13, ncols=4, figsize=(30, 40))
            for idx, feature in enumerate(data.columns[1:]):
              data.plot(feature, 'machine_status', subplots=True, kind="scatter", ax=axes[idx // 4, idx % 4])
            st.write(fig)
        elif request == "Диаграмма рассеивания двух признаков":
            x_axis = st.selectbox("Выберите столбец для x-axis", data.columns)
            y_axis = st.selectbox("Выберите столбец для y-axis", data.columns)
            visualize_data(data, x_axis, y_axis)
        elif request == "Круговая диаграмма":
            labelss = st.selectbox("Выберите столбец", ["month", "machine_status"])
            visual(data, labelss)
    elif page == "Информация о построенных моделях":
        st.title("Информация о моделях")
        st.write("Для перехода на иную страницу можете выбрать ее слева")
        st.markdown("""Было обучено три модели для классификации на отбалансированных данных с отобранными признаками:
* LogisticRegressor - данная модель использует всего эти три признака: sensor04, sensor10, sensor12. Она показала достаточно хорошие результаты и при этом, на мой взгляд, не является переобученной.
Здесь представлен отчет того, как справлялась на тестовой выборке данная модель

            classes    precision   recall   f1-score   support
           0       1.00      0.98      0.99     30986
           1       0.87      0.95      0.91     30782
           2       0.93      0.86      0.89     30859

    accuracy: 0.93
   macro avg: 0.93
weighted avg: 0.93

* GaussianNB - модель обучалась на тех же трех признаках. Ее результаты:

            classes   precision    recall  f1-score   support
           0       0.87      0.97      0.92     30986
           1       0.87      0.99      0.92     30782
           2       0.95      0.71      0.82     30859

    accuracy: 0.89
   macro avg: 0.89
weighted avg: 0.89

Целевой признак:
* BaggingClassifier - ансамблевая модель, которая справилась лучше всего, однако нельзя с уверенностью сказать, что она не является переобученной. Также по результатам она обгоняет предыдущие две:

           classes  precision    recall  f1-score   support
           0       1.00      1.00      1.00     30986
           1       1.00      1.00      1.00     30782
           2       1.00      1.00      1.00     30859

    accuracy: 1.00
   macro avg: 1.00
weighted avg: 1.00

        """)
    elif page == "Запрос к модели":
        st.title("Запрос к модели")
        st.write("Для перехода на иную страницу можете выбрать ее слева")

        sensor4 = st.number_input("Задайте значение на 4-м датчике. Минимальное значение 3, максимальное 800:", 3, 800, 10)
        sensor4 = float(sensor4)

        sensor10 = st.slider("Задайте значение на 10-м датчике: ", min_value=0, max_value=75, value=1, step=1)
        sensor10 = float(sensor10)

        sensor12 = st.slider("Задайте значение на 12-м датчике: ", min_value=0, max_value=45, value=1, step=1)
        sensor12 = float(sensor12)

        if st.button('Получить предсказание'):
            frame = [sensor4, sensor10, sensor12]
            frame = np.array(frame).reshape((1, -1))
            data_df = pd.DataFrame(frame)
            pred1 = model_1.predict(data_df)
            st.write(f"Значение, предсказанное с помощью модели Логистической регресиии: {pred1[0]:.2f}, точность ответа: 0.93")
        else:
            pass



@st.cache_data
def visual(test_data, labelss):
    fig, ax = plt.subplots()
    test_data.groupby(labelss).size().plot(kind='pie')
    plt.show()
    st.write(fig)
def visualize_data(test_data, x_axis, y_axis):
    fig, ax = plt.subplots(figsize=(16,14))
    ax.scatter(x = test_data[x_axis], y = test_data[y_axis])
    plt.xlabel(x_axis)
    plt.ylabel(y_axis)
    plt.show()
    st.write(fig)

def load_model(path_to_file):
    with open(path_to_file, 'rb') as model_file:
        model = joblib.load(model_file)
    return model

@st.cache_data
def load_test_data(path_to_file):
    df = pd.read_csv(path_to_file, index_col='Unnamed: 0')
    df = df.drop(labels=['machine_status'], axis=1)
    return df
def load_data(path_to_file):
    df = pd.read_csv(path_to_file, index_col='Unnamed: 0')
    return df
if __name__ == "__main__":
    main()
```
