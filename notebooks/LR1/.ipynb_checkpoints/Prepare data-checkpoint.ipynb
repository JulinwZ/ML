```python
import pandas as pd
```

```python
df = pd.read_csv('mumbai_houses_task.csv', sep = ',')
```

```python
df
```

```python
pd.isnull(df).sum()
```

```python
temp = df['Status'].dropna().mode()[0]
df['Status'] = df['Status'].fillna(temp)

temp = df['Furnished_status'].dropna().mode()[0]
df['Furnished_status'] = df['Furnished_status'].fillna(temp)

df
```

```python
pd.isnull(df).sum()
```

```python
one_hot = pd.get_dummies(df['Status'])
df = df.drop(columns = ['Status'])
df = df.join(one_hot)

one_hot = pd.get_dummies(df['neworold'])
df = df.drop(columns = ['neworold'])
df = df.join(one_hot)

one_hot = pd.get_dummies(df['Furnished_status'])
df = df.drop(columns = ['Furnished_status'])
df = df.join(one_hot)

one_hot = pd.get_dummies(df['type_of_building'])
df = df.drop(columns = ['type_of_building'])
df = df.join(one_hot)
```

```python
df
```

```python
df.to_csv('result.csv')
```
